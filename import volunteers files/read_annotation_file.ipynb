```python
import json

data_file = "./downloaded-done/HawaiiBeetles07Marija.json"

with open(data_file, "r") as f:
    data = json.load(f)

print(data[0]["created_at"])
```

```python
all_update_dates = {}

for item in data:
    date = item["updated_at"]
    date = date.split("T")[0] + "T" + date.split("T")[1].split(":")[0] # Extract the date and hour part (YYYY-MM-DD-HH)
    if date in all_update_dates:
        all_update_dates[date] += 1
    else:
        all_update_dates[date] = 1

print(all_update_dates)
```

```python
import matplotlib.pyplot as plt
X = list(all_update_dates.keys())
X.sort()
Y = [all_update_dates[x] for x in X]
plt.bar(X, Y)
plt.xticks(rotation=45)
```
