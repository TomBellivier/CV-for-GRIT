```python
from pathlib import Path    
import os
BASE_PATH = Path("D:/ORBIT bees/drive images/")
SAVE_PATH = Path("D:/ORBIT bees/selected_images/")

os.makedirs(SAVE_PATH, exist_ok=True)

all_files = list(BASE_PATH.rglob("*.jpg"))
print(len(all_files))
```

```python
kept_images = [f for f in all_files if "_001" in f.stem]
print(len(kept_images))
```

```python
import PIL.Image as Image
import tqdm

for img_file in tqdm.tqdm(kept_images):
    img = Image.open(img_file)
    img.save(SAVE_PATH / img_file.name)
    
```
